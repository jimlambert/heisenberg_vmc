```python
import numpy as np
import matplotlib.pyplot as plt
import AnalysisUtls as an
from statistics import mean, stdev

binSize=100
n=50
fs=20
energyArray=an.ReadEnergies("equildata-1-2.dat")
aveEnergies=an.AverageEnergies(energyArray, binSize)
engX=np.arange(0, len(energyArray))
aveX=np.arange(0, len(energyArray), binSize)
plt.figure(figsize=(20, 10))
plt.plot(engX[::n], energyArray[::n], '-b', alpha=0.5)
plt.plot(aveX, aveEnergies, 'or')
plt.xlim(min(engX), max(engX))
plt.show()
```
