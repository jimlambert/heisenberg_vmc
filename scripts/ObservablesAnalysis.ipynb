```python
import AnalysisUtls as an
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
files = "../n4_optvals_ising*"
header, obsAve, obsErr = an.ProcessDataFiles(files, True)
```

```python
fs = 30
ls = 20
plt.figure(figsize=(20,10))
#plt.plot(np.arange(0, len(obsAve[0])), obsAve[0], '.')
plt.errorbar(np.arange(0, len(obsAve[0])), obsAve[0], yerr=obsErr[0], marker="d")
plt.xlabel("variational step", fontsize=fs)
plt.ylabel("$<E>$", fontsize=fs)
plt.title("Heisenberg Energy, N=8", fontsize=fs)
plt.axhline(y=-1.5)
plt.tick_params(axis='both', which='major', labelsize=ls)
plt.savefig("n8heis_energy.pdf")
plt.show()
```
