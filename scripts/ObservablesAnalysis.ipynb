```python
import AnalysisUtls as an
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
files = "../../vmc-dat/n4obsvals*"
header, obsAve, obsErr = an.ProcessDataFiles(files, True)
```

```python
fs = 30
ls = 20
plt.figure(figsize=(20,10))
#plt.plot(np.arange(0, len(obsAve[0])), obsAve[0], '.')
plt.errorbar(np.arange(0, len(obsAve[0])), obsAve[0], yerr=obsErr[0], marker="d")
plt.xlabel("variational step", fontsize=fs)
plt.ylabel("$<E>$", fontsize=fs)
plt.title(header[0]+" $N$=10", fontsize=fs)
plt.axhline(y=-5.317)
plt.tick_params(axis='both', which='major', labelsize=ls)
#plt.savefig("n10-h1.0-enrg_ising.pdf")
plt.show()
print(header)
```

```python
index=1
plt.figure(figsize=(20,10))
#plt.plot(np.arange(0, len(obsAve[0])), obsAve[0], '.')
plt.errorbar(np.arange(0, len(obsAve[index])), obsAve[index], yerr=obsErr[index], marker="d")
plt.xlabel("variational step", fontsize=fs)
plt.ylabel("$<E>$", fontsize=fs)
plt.title(header[index], fontsize=fs)
plt.tick_params(axis='both', which='major', labelsize=ls)
plt.show()
```
