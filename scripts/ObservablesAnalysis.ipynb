```python
import AnalysisUtls as an
import numpy as np
import matplotlib.pyplot as plt
import glob

files = "../../dat/n10/n10_optvals*"
nfiles="../n10_optvals*"
header, obsAve, obsErr = an.ProcessDataFiles(files, True)
nheader, nobsAve, nobsErr = an.ProcessDataFiles(nfiles, True)
```

```python
fs = 30
ls = 20
plt.figure(figsize=(20,10))
#plt.plot(np.arange(0, len(obsAve[0])), obsAve[0], '.')
plt.errorbar(np.arange(0, len(obsAve[0])), obsAve[0], yerr=obsErr[0], marker="d")
plt.xlabel("variational step", fontsize=fs)
plt.ylabel("$<E>$", fontsize=fs)
plt.tick_params(axis='both', which='major', labelsize=ls)
plt.savefig("n10energy.pdf")
plt.show()
```

```python
ls=20
fs=30

head, data = an.ReadParamsDat("../var_params10_trial1.dat")

nsteps = np.arange(0, len(data[0]))
plt.figure(figsize=(20,20))
plt.subplot(221)
plt.plot(nsteps, data[0], '.')
plt.title(head[0]+" N=10", fontsize=fs)
plt.ylabel("$\mu_0$", fontsize=ls)
plt.xlabel("optimization steps", fontsize=ls)
plt.tick_params(axis='both', which='major', labelsize=ls)
plt.subplot(222)
plt.plot(nsteps, data[1], '.')
plt.title(head[1]+" N=10", fontsize=fs)
plt.ylabel("$t_{i,i+1}$", fontsize=ls)
plt.xlabel("optimization steps", fontsize=ls)
plt.tick_params(axis='both', which='major', labelsize=ls)
plt.subplot(223)
plt.plot(nsteps, data[2], '.')
plt.title(head[2]+" N=10", fontsize=fs)
plt.ylabel("$\Delta_{ii}$", fontsize=ls)
plt.xlabel("optimization steps", fontsize=ls)
plt.tick_params(axis='both', which='major', labelsize=ls)
plt.subplot(224)
plt.plot(nsteps, data[3], '.')
plt.title(head[3]+" N=10", fontsize=fs)
plt.ylabel("$\Delta_{ii+1}$", fontsize=ls)
plt.xlabel("optimization steps", fontsize=ls)
plt.tick_params(axis='both', which='major', labelsize=ls)
plt.tight_layout()
plt.savefig("n10varparams.pdf")
plt.show()
```
