```python
import AnalysisUtls as an
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
files = "../../dat/n4_optvals_ising*"
header, obsAve, obsErr = an.ProcessDataFiles(files, True)
```

```python
fs = 30
ls = 20
plt.figure(figsize=(20,10))
#plt.plot(np.arange(0, len(obsAve[0])), obsAve[0], '.')
plt.errorbar(np.arange(0, len(obsAve[0])), obsAve[0], yerr=obsErr[0], marker="d")
plt.xlabel("variational step", fontsize=fs)
plt.ylabel("$<E>$", fontsize=fs)
plt.title("Heisenberg Energy", fontsize=fs)
plt.axhline(y=-2.801)
plt.tick_params(axis='both', which='major', labelsize=ls)
plt.savefig("n10energy_ising.pdf")
plt.show()
```

```python
for i in range(0, len(obsAve[0])-1):
    print(i, ":", '\t', obsAve[0][i+1]-obsAve[0][i], '\n')
```
