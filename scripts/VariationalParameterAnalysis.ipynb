```python
import numpy as np
import matplotlib.pyplot as plt
import AnalysisUtls as an
from statistics import mean, stdev

ls=20
fs=30

head, data = an.ReadParamsDat("../var_params.dat")

nsteps = np.arange(0, len(data[0]))
plt.figure(figsize=(20,10))
plt.subplot(131)
plt.plot(nsteps, data[0], '.')
plt.title(head[0]+" N=10", fontsize=fs)
plt.ylabel("$\mu_0$", fontsize=ls)
plt.xlabel("optimization steps", fontsize=ls)
plt.subplot(132)
plt.plot(nsteps, data[1], '.')
plt.title(head[1]+" N=10", fontsize=fs)
plt.ylabel("$t_{i,i+1}$", fontsize=ls)
plt.xlabel("optimization steps", fontsize=ls)
plt.subplot(133)
plt.plot(nsteps, data[2], '.')
plt.title(head[2]+" N=10", fontsize=fs)
plt.ylabel("$\Delta_{ii}$", fontsize=ls)
plt.xlabel("optimization steps", fontsize=ls)
plt.savefig("n10.pdf")
plt.show()
```
